### Line graph creation 

##### Object hierarchy 

- Figure Object (the entire board)
- Axis Object (contains the x and y axis and is in the context of Figure object)
    -- It is a container object for the markers, lines and other graphical elements
fig.add_subplot(5,1,1)
Returns an axis object

- One Figure object can contain multiple grids (row x columns)
- Each cell in the grid can contain a axis object(or a sub-plot)


```python
import pandas as p
import matplotlib.pyplot as pt


unrate = p.read_csv("unrate.csv")

unrate["DATE"] = p.to_datetime(unrate["DATE"])

unrate["MONTH"] = unrate["DATE"].dt.month

fig = pt.figure(figsize=(10,6))
colors = ['red', 'blue', 'green', 'orange', 'black']
a=1948
for i in range(5):
    b=a+i
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    
    pt.plot(subset['MONTH'], subset['UNRATE'], c=colors[i], label=str(b))

pt.legend(loc='upper left')
pt.title("Monthly Unemployment Trends, 1948-1952")
pt.xlabel("Month, Integer")
pt.ylabel("Unemployment Rate, Percent")
pt.show()
```

```python

import matplotlib.pyplot as plt
from numpy import arange

# Returns one figure and one axis object
fig,ax = plt.subplots()

num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']

bar_heights = norm_reviews[num_cols].iloc[0].values

# numpy.arange() fnction similar to range
bar_positions = arange(5) + 0.75

ax.bar(bar_positions, bar_heights, 0.5)
#Set the tick positions, labels and label allignment using axis object
ax.set_xticks(range(1,6))
ax.set_xticklabels(num_cols,rotation=90)
ax.set_xlabel("Rating Source")
ax.set_ylabel("Average Rating")
plt.title("Average User Rating For Avengers: Age of Ultron (2015)")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

reviews = pd.read_csv('fandango_score_comparison.csv')

cols = ['FILM', 'RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue']
norm_reviews = reviews[cols]
#print(norm_reviews[:5])

fandango_distribution=norm_reviews["Fandango_Ratingvalue"].value_counts().sort_index()

imdb_distribution=norm_reviews["IMDB_norm"].value_counts().sort_index()

print(fandango_distribution)
print(imdb_distribution)
```

#### Frequency Distribution : 
Counts(occurence) of unique values in the given set of data

- Can be achieved using Series.value_counts()

To compare the frequency distributions of the 2 sets, the sets can be divided into multiple equilength ranges (with max and min value be the max and min value of union of 2 sets). These ranges are called as bins. A histogram can be created to compare the value change over these bins

```python
fig = plt.figure(figsize=(20,40))
ax1 = fig.add_subplot(4,1,1)
ax2 = fig.add_subplot(4,1,2)
ax3 = fig.add_subplot(4,1,3)
ax4 = fig.add_subplot(4,1,4)

ax1.hist(norm_reviews["Fandango_Ratingvalue"], bins=20,range=(1,5))
ax1.set_title("Distribution of Fandango Ratings")
ax1.set_ylim(0,50)
ax2.hist(norm_reviews["RT_user_norm"], bins=20,range=(1,5))
ax2.set_title("Distribution of Rotten Tomatoes Ratings")
ax2.set_ylim(0,50)
ax3.hist(norm_reviews["Metacritic_user_nom"], bins=20,range=(1,5))
ax3.set_title("Distribution of Metacritic Ratings")
ax3.set_ylim(0,50)
ax4.hist(norm_reviews["IMDB_norm"], bins=20,range=(1,5))
ax4.set_title("Distribution of IMDB Ratings")
ax4.set_ylim(0,50)

plt.show()
```
